```python
# ============================================================
# ESG RATING PREDICTION — XGBoost + SHAP
# Correcciones metodológicas:
#   - Split train/test ANTES de imputar
#   - Imputación aprendida solo en train
#   - One-hot encoding aprendido solo en train
#   - SMOTE-ENN solo en train
#   - SHAP feature selection solo en train
#   - Métricas adicionales: balanced accuracy, macro F1
# CAMBIO v5-final:
#   - Modelo A usa smoteenn_soft (ENN con kind_sel='mode')
#     para evitar eliminación excesiva de muestras
#   - Modelo B mantiene smoteenn_strict (comportamiento original)
# ============================================================

!pip install yfinance imbalanced-learn xgboost shap --quiet

# ══════════════════════════════════════════════════════════════
# BLOQUE 1 — CARGA Y ENRIQUECIMIENTO CON YFINANCE
# ══════════════════════════════════════════════════════════════

from google.colab import drive
drive.mount('/content/drive')

import os
import time
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import yfinance as yf

RANDOM_STATE = 42

RESULTS_DIR = "/content/drive/MyDrive/Colab Notebooks/XAI-ESG/results_v5"
os.makedirs(RESULTS_DIR, exist_ok=True)

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/XAI-ESG/SP 500 ESG Risk Ratings.csv')

print(f"Dataset original: {df.shape[0]} empresas")
print("\nDistribución original de clases:")
print(df['ESG Risk Level'].value_counts())


def obtener_finanzas(ticker):
    try:
        info = yf.Ticker(ticker).info
        return {
            'Symbol': ticker,
            # Financieras
            'Market_Cap': info.get('marketCap', None),
            'EBITDA': info.get('ebitda', None),
            'Total_Revenue': info.get('totalRevenue', None),
            'Profit_Margin': info.get('profitMargins', None),
            'ROE': info.get('returnOnEquity', None),
            'Debt_to_Equity': info.get('debtToEquity', None),
            'PE_Ratio': info.get('trailingPE', None),
            'Beta': info.get('beta', None),
            # Gobernanza vía Yahoo/ISS
            'Audit_Risk': info.get('auditRisk', None),
            'Board_Risk': info.get('boardRisk', None),
            'Compensation_Risk': info.get('compensationRisk', None),
            'Shareholder_Rights_Risk': info.get('shareHolderRightsRisk', None),
            'Num_Directors': info.get('numberOfDirectors', None),
            'Held_Pct_Insiders': info.get('heldPercentInsiders', None),
            'Held_Pct_Institutions': info.get('heldPercentInstitutions', None),
        }
    except Exception:
        return {
            'Symbol': ticker,
            'Market_Cap': None, 'EBITDA': None, 'Total_Revenue': None,
            'Profit_Margin': None, 'ROE': None, 'Debt_to_Equity': None,
            'PE_Ratio': None, 'Beta': None, 'Audit_Risk': None,
            'Board_Risk': None, 'Compensation_Risk': None,
            'Shareholder_Rights_Risk': None, 'Num_Directors': None, 'Held_Pct_Insiders': None,
            'Held_Pct_Institutions': None,
        }


tickers = df['Symbol'].dropna().unique()
print(f"\nDescargando datos financieros para {len(tickers)} empresas...")
datos = []

for i, t in enumerate(tickers):
    if i % 50 == 0 and i > 0:
        print(f"  {i}/{len(tickers)}...")
    datos.append(obtener_finanzas(t))
    time.sleep(0.1)

df_fin = pd.DataFrame(datos)
df = pd.merge(df, df_fin, on='Symbol', how='left')
print(f"\nDataset enriquecido: {df.shape}")


# ══════════════════════════════════════════════════════════════
# EXPORT — TABLA MAESTRA ENRIQUECIDA
# ══════════════════════════════════════════════════════════════

from sklearn.model_selection import train_test_split

df_master = df.copy()

mapeo_3 = {'Negligible': 0, 'Low': 0, 'Medium': 1, 'High': 2, 'Severe': 2}
mapeo_2 = {'Negligible': 0, 'Low': 0, 'Medium': 0, 'High': 1, 'Severe': 1}
nombres_3_dict = {0: 'Riesgo Bajo', 1: 'Riesgo Medio', 2: 'Riesgo Alto'}
nombres_2_dict = {0: 'No-Riesgo (Bajo/Medio)', 1: 'Riesgo (Alto/Severo)'}

df_master["target_3clases"] = df_master["ESG Risk Level"].map(mapeo_3)
df_master["target_3clases_label"] = df_master["target_3clases"].map(nombres_3_dict)
df_master["target_binario"] = df_master["ESG Risk Level"].map(mapeo_2)
df_master["target_binario_label"] = df_master["target_binario"].map(nombres_2_dict)
df_master["target_valido"] = df_master["ESG Risk Level"].notna()

df_valid = df_master[df_master["target_binario"].notna()].copy()
train_idx, test_idx = train_test_split(
    df_valid.index, test_size=0.2, random_state=42,
    stratify=df_valid["target_binario"]
)
df_master["split_modelo_binario"] = "sin_target"
df_master.loc[train_idx, "split_modelo_binario"] = "train"
df_master.loc[test_idx, "split_modelo_binario"] = "test"

df_valid_3 = df_master[df_master["target_3clases"].notna()].copy()
train_idx_3, test_idx_3 = train_test_split(
    df_valid_3.index, test_size=0.2, random_state=42,
    stratify=df_valid_3["target_3clases"]
)
df_master["split_modelo_3clases"] = "sin_target"
df_master.loc[train_idx_3, "split_modelo_3clases"] = "train"
df_master.loc[test_idx_3, "split_modelo_3clases"] = "test"

columnas_prioritarias = [
    "Symbol", "Name", "Address", "Sector", "Industry", "Full Time Employees",
    "Description", "ESG Risk Level", "target_3clases", "target_3clases_label",
    "target_binario", "target_binario_label", "split_modelo_3clases",
    "split_modelo_binario", "Total ESG Risk score", "Environment Risk Score",
    "Social Risk Score", "Governance Risk Score", "ESG Risk Percentile",
    "Controversy Level", "Controversy Score", "Market_Cap", "EBITDA",
    "Total_Revenue", "Profit_Margin", "ROE", "Debt_to_Equity", "PE_Ratio",
    "Beta", "Audit_Risk", "Board_Risk", "Compensation_Risk",
    "Shareholder_Rights_Risk", "Num_Directors",
    "Held_Pct_Insiders", "Held_Pct_Institutions"
]
columnas_prioritarias = [c for c in columnas_prioritarias if c in df_master.columns]
otras_columnas = [c for c in df_master.columns if c not in columnas_prioritarias]
df_master = df_master[columnas_prioritarias + otras_columnas]

ruta_csv = os.path.join(RESULTS_DIR, "dataset_maestro_empresas_variables.csv")
ruta_excel = os.path.join(RESULTS_DIR, "dataset_maestro_empresas_variables.xlsx")
df_master.to_csv(ruta_csv, index=False)
df_master.to_excel(ruta_excel, index=False)
print("Dataset maestro guardado.")
print("\nDimensión dataset maestro:", df_master.shape)
display(df_master.head())


# ══════════════════════════════════════════════════════════════
# BLOQUE 2 — DEFINICIÓN DE VARIABLES Y LIMPIEZA SIN LEAKAGE
# ══════════════════════════════════════════════════════════════

from collections import Counter

vars_yfinance = [
    'Market_Cap', 'EBITDA', 'Total_Revenue', 'Profit_Margin',
    'ROE', 'Debt_to_Equity', 'PE_Ratio', 'Beta',
    'Audit_Risk', 'Board_Risk', 'Compensation_Risk',
    'Shareholder_Rights_Risk',
    'Num_Directors', 'Held_Pct_Insiders', 'Held_Pct_Institutions',
]

columnas_descartadas = [
    'Symbol', 'Name', 'Address', 'Description',
    'Environment Risk Score', 'Social Risk Score', 'Governance Risk Score',
    'ESG Risk Percentile', 'Controversy Level', 'Controversy Score',
    'Total ESG Risk score'
]

df['Full Time Employees'] = (
    df['Full Time Employees']
    .astype(str)
    .str.replace(',', '', regex=False)
    .replace('nan', np.nan)
    .astype(float)
)

df_model = df.drop(columns=[c for c in columnas_descartadas if c in df.columns], errors='ignore')
df_model = df_model.dropna(subset=['ESG Risk Level'])

print("\nColumnas disponibles tras eliminar leakage:")
print(df_model.columns.tolist())
print("\nDistribución tras limpieza inicial:")
print(df_model['ESG Risk Level'].value_counts())


# ══════════════════════════════════════════════════════════════
# BLOQUE 3 — TARGETS: MODELO A Y MODELO B
# ══════════════════════════════════════════════════════════════

nombres_3 = ['Riesgo Bajo', 'Riesgo Medio', 'Riesgo Alto']
nombres_2 = ['No-Riesgo (Bajo/Medio)', 'Riesgo (Alto/Severo)']

y_3_full = df_model['ESG Risk Level'].map(mapeo_3)
y_2_full = df_model['ESG Risk Level'].map(mapeo_2)

mask_valid = y_3_full.notna() & y_2_full.notna()
df_model = df_model.loc[mask_valid].copy()
y_3_full = y_3_full.loc[mask_valid].astype(int).values
y_2_full = y_2_full.loc[mask_valid].astype(int).values

X_full = df_model.drop(columns=['ESG Risk Level'])

print("\nDistribución Modelo A (3 clases):", Counter(y_3_full))
print("Distribución Modelo B (2 clases):", Counter(y_2_full))
print("\nDimensión X_full:", X_full.shape)


# ══════════════════════════════════════════════════════════════
# BLOQUE 4 — PREPROCESAMIENTO SIN LEAKAGE
# ══════════════════════════════════════════════════════════════

def preparar_train_test_sin_leakage(X_train_raw, X_test_raw):
    X_train = X_train_raw.copy()
    X_test = X_test_raw.copy()

    for col in ['Sector', 'Industry']:
        if col not in X_train.columns:
            X_train[col] = 'Unknown'
            X_test[col] = 'Unknown'
    X_train['Sector'] = X_train['Sector'].fillna('Unknown').astype(str)
    X_test['Sector'] = X_test['Sector'].fillna('Unknown').astype(str)
    X_train['Industry'] = X_train['Industry'].fillna('Unknown').astype(str)
    X_test['Industry'] = X_test['Industry'].fillna('Unknown').astype(str)

    candidate_num_cols = [c for c in X_train.columns if c not in ['Sector', 'Industry']]
    for col in candidate_num_cols:
        X_train[col] = pd.to_numeric(X_train[col], errors='coerce')
        X_test[col] = pd.to_numeric(X_test[col], errors='coerce')

    cols_all_nan_train = [c for c in candidate_num_cols if X_train[c].isna().all()]
    if cols_all_nan_train:
        print("\nColumnas eliminadas por ser 100% NaN en TRAIN:")
        for c in cols_all_nan_train:
            print(f" - {c}")
        X_train = X_train.drop(columns=cols_all_nan_train)
        X_test = X_test.drop(columns=cols_all_nan_train)
        candidate_num_cols = [c for c in candidate_num_cols if c not in cols_all_nan_train]

    for col in candidate_num_cols:
        sector_medians = X_train.groupby('Sector')[col].median()
        global_median = X_train[col].median()
        if pd.isna(global_median):
            X_train = X_train.drop(columns=[col])
            X_test = X_test.drop(columns=[col])
            continue
        X_train[col] = X_train[col].fillna(X_train['Sector'].map(sector_medians)).fillna(global_median)
        X_test[col] = X_test[col].fillna(X_test['Sector'].map(sector_medians)).fillna(global_median)

    X_train_enc = pd.get_dummies(X_train, columns=['Sector', 'Industry'], drop_first=True)
    X_test_enc = pd.get_dummies(X_test, columns=['Sector', 'Industry'], drop_first=True)
    X_test_enc = X_test_enc.reindex(columns=X_train_enc.columns, fill_value=0)
    X_train_enc = X_train_enc.astype(float)
    X_test_enc = X_test_enc.astype(float)

    return X_train_enc, X_test_enc


# ══════════════════════════════════════════════════════════════
# BLOQUE 5 — FUNCIÓN AUXILIAR COMPLETA
# ══════════════════════════════════════════════════════════════

from imblearn.combine import SMOTEENN
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import EditedNearestNeighbours
from sklearn.model_selection import RandomizedSearchCV
from sklearn.dummy import DummyClassifier
from sklearn.metrics import (
    accuracy_score, balanced_accuracy_score, classification_report,
    roc_auc_score, confusion_matrix, f1_score
)
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns


def calcular_importancia_shap(shap_values):
    if isinstance(shap_values, list):
        return np.sum([np.abs(v).mean(axis=0) for v in shap_values], axis=0)
    elif len(shap_values.shape) == 3:
        return np.abs(shap_values).mean(axis=(0, 2))
    else:
        return np.abs(shap_values).mean(axis=0)


# ── ÚNICO CAMBIO respecto al código original: ────────────────
# Se añade el parámetro 'estrategia_balanceo' a la función.
# Modelo A lo llamará con 'smoteenn_soft' para evitar la
# eliminación excesiva de muestras que ocurría con ENN estricto
# (de 344 muestras pasaba a solo 79).
# Modelo B lo llamará con 'smoteenn_strict' (comportamiento
# original, que funcionaba bien: de 344 pasaba a 333).
# ─────────────────────────────────────────────────────────────
def pipeline_completo_v5(X_raw, y, nombres_clases, titulo_modelo,
                          nombre_archivo,
                          estrategia_balanceo='smoteenn_strict'):
    """
    estrategia_balanceo:
      'smoteenn_strict' — SMOTE + ENN agresivo (original, para Modelo B)
      'smoteenn_soft'   — SMOTE + ENN suave kind_sel='mode' (para Modelo A)
      'smote_only'      — solo oversample, sin limpieza (fallback)
    """

    n_clases = len(nombres_clases)
    es_binario = (n_clases == 2)

    print("\n" + "="*80)
    print(f"  {titulo_modelo}")
    print("="*80)

    # ── 1. Split estratificado ANTES de imputar ──────────────
    X_train_raw, X_test_raw, y_train_raw, y_test = train_test_split(
        X_raw, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
    )
    print(f"\nTrain raw: {len(X_train_raw)} | Test raw: {len(X_test_raw)}")
    print("Distribución train raw:", Counter(y_train_raw))
    print("Distribución test:", Counter(y_test))

    # ── 2. Preprocesamiento aprendido SOLO en train ──────────
    X_train_enc, X_test_enc = preparar_train_test_sin_leakage(X_train_raw, X_test_raw)
    print("\nDimensiones tras preprocesamiento:")
    print("X_train_enc:", X_train_enc.shape)
    print("X_test_enc: ", X_test_enc.shape)

    # ── 3. Baseline simple ───────────────────────────────────
    dummy = DummyClassifier(strategy='most_frequent', random_state=RANDOM_STATE)
    dummy.fit(X_train_enc, y_train_raw)
    y_dummy = dummy.predict(X_test_enc)
    acc_dummy = accuracy_score(y_test, y_dummy)
    balacc_dummy = balanced_accuracy_score(y_test, y_dummy)
    print("\nBaseline DummyClassifier:")
    print(f"Accuracy:          {acc_dummy:.4f}")
    print(f"Balanced Accuracy: {balacc_dummy:.4f}")

    # ── 4. SMOTE-ENN SOLO en training ────────────────────────
    # estrategia_balanceo controla la agresividad de ENN:
    #   - Modelo A usa 'smoteenn_soft': ENN elimina solo si la
    #     MAYORÍA de vecinos discrepa (kind_sel='mode'), mucho
    #     menos agresivo. Evita pasar de 344 a 79 muestras.
    #   - Modelo B usa 'smoteenn_strict': comportamiento por
    #     defecto, que en este caso funcionaba bien (344→333).
    min_class_count = min(Counter(y_train_raw).values())

    if min_class_count <= 2:
        print("\n⚠️  Clase minoritaria muy pequeña — sin balanceo.")
        X_train_bal = X_train_enc.copy()
        y_train_bal = y_train_raw.copy()
    else:
        k_neighbors = min(5, min_class_count - 1)
        smote = SMOTE(k_neighbors=k_neighbors, random_state=RANDOM_STATE)

        if estrategia_balanceo == 'smote_only':
            X_train_bal_array, y_train_bal = smote.fit_resample(
                X_train_enc, y_train_raw
            )
        elif estrategia_balanceo == 'smoteenn_soft':
            enn = EditedNearestNeighbours(kind_sel='mode', n_neighbors=3)
            smote_enn = SMOTEENN(smote=smote, enn=enn, random_state=RANDOM_STATE)
            X_train_bal_array, y_train_bal = smote_enn.fit_resample(
                X_train_enc, y_train_raw
            )
        else:  # smoteenn_strict — comportamiento original
            smote_enn = SMOTEENN(smote=smote, random_state=RANDOM_STATE)
            X_train_bal_array, y_train_bal = smote_enn.fit_resample(
                X_train_enc, y_train_raw
            )

        X_train_bal = pd.DataFrame(X_train_bal_array, columns=X_train_enc.columns)

    print(f"\nTrain tras balanceo ({estrategia_balanceo}):")
    print(Counter(y_train_bal))
    print("Dimensión:", X_train_bal.shape)

    # ── 5. XGBoost inicial para selección SHAP ───────────────
    objective = 'binary:logistic' if es_binario else 'multi:softprob'

    modelo_ini = xgb.XGBClassifier(
        objective=objective,
        random_state=RANDOM_STATE,
        eval_metric='logloss' if es_binario else 'mlogloss',
        n_estimators=200, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8
    )
    modelo_ini.fit(X_train_bal, y_train_bal)

    explainer_ini = shap.TreeExplainer(modelo_ini)
    sv_ini = explainer_ini.shap_values(X_train_enc)
    imp = calcular_importancia_shap(
        np.array(sv_ini) if not isinstance(sv_ini, list) else sv_ini
    )

    df_imp = (
        pd.DataFrame({'Variable': X_train_enc.columns, 'Importancia': imp})
        .sort_values('Importancia', ascending=False)
    )
    top15 = df_imp.head(15)['Variable'].tolist()

    print("\nTop 15 variables seleccionadas por SHAP inicial:")
    print(df_imp.head(15).to_string(index=False))

    df_imp.to_csv(
        os.path.join(RESULTS_DIR, f"{nombre_archivo}_shap_importancia_inicial.csv"),
        index=False
    )

    plt.figure(figsize=(10, 7))
    shap.summary_plot(sv_ini, X_train_enc, plot_type="bar", show=False)
    plt.title(f"SHAP Inicial — {titulo_modelo}", pad=12)
    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_DIR, f"{nombre_archivo}_shap_inicial.png"), dpi=200)
    plt.show()

    # ── 6. Modelo final con Top 15 + RandomizedSearchCV ──────
    X_tr15 = X_train_bal[top15]
    X_te15 = X_test_enc[top15]

    params = {
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [3, 4, 5, 6],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.5],
        'min_child_weight': [1, 3, 5],
    }
    base = xgb.XGBClassifier(
        objective=objective, random_state=RANDOM_STATE,
        eval_metric='logloss' if es_binario else 'mlogloss'
    )
    search = RandomizedSearchCV(
        estimator=base, param_distributions=params,
        n_iter=30, scoring='f1_macro', cv=3,
        random_state=RANDOM_STATE, n_jobs=-1, verbose=0
    )
    search.fit(X_tr15, y_train_bal)
    modelo_fin = search.best_estimator_

    print("\nMejores hiperparámetros:")
    print(search.best_params_)

    # ── 7. Evaluación en TEST intacto ────────────────────────
    y_pred = modelo_fin.predict(X_te15)
    y_proba = modelo_fin.predict_proba(X_te15)

    acc = accuracy_score(y_test, y_pred)
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1_macro = f1_score(y_test, y_pred, average='macro')
    f1_weighted = f1_score(y_test, y_pred, average='weighted')

    try:
        auc = roc_auc_score(y_test, y_proba[:, 1]) if es_binario else \
              roc_auc_score(y_test, y_proba, multi_class='ovr', average='weighted')
    except Exception as e:
        auc = np.nan
        print(f"AUC no calculable: {e}")

    print("\nRESULTADOS EN TEST")
    print(f"Accuracy:          {acc:.4f}")
    print(f"Balanced Accuracy: {bal_acc:.4f}")
    print(f"F1 macro:          {f1_macro:.4f}")
    print(f"F1 weighted:       {f1_weighted:.4f}")
    print(f"AUC:               {auc:.4f}" if not np.isnan(auc) else "AUC: N/A")
    print("\nClassification report:")
    print(classification_report(y_test, y_pred, target_names=nombres_clases, digits=3))

    metricas = pd.DataFrame([{
        'modelo': titulo_modelo,
        'estrategia_balanceo': estrategia_balanceo,
        'accuracy': acc, 'balanced_accuracy': bal_acc,
        'f1_macro': f1_macro, 'f1_weighted': f1_weighted, 'auc': auc,
        'dummy_accuracy': acc_dummy, 'dummy_balanced_accuracy': balacc_dummy,
        'n_train_raw': len(X_train_raw), 'n_test': len(X_test_raw),
        'n_train_after_smoteenn': len(X_train_bal)
    }])
    metricas.to_csv(
        os.path.join(RESULTS_DIR, f"{nombre_archivo}_metricas.csv"), index=False
    )

    # ── 8. Matriz de confusión ───────────────────────────────
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(7, 5))
    sns.set_theme(font_scale=1.1, style="white")
    sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False,
                xticklabels=nombres_clases, yticklabels=nombres_clases,
                linewidths=1, linecolor='black')
    plt.title(f'Matriz de Confusión — {titulo_modelo}', pad=15, fontsize=13, fontweight='bold')
    plt.xlabel('Clase Predicha', labelpad=8, fontweight='bold')
    plt.ylabel('Clase Real', labelpad=8, fontweight='bold')
    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_DIR, f"{nombre_archivo}_confusion_matrix.png"), dpi=200)
    plt.show()

    # ── 9. SHAP final sobre TEST ─────────────────────────────
    explainer_fin = shap.TreeExplainer(modelo_fin)
    expl_obj = explainer_fin(X_te15)
    sv_fin = explainer_fin.shap_values(X_te15)

    plt.figure(figsize=(10, 6))
    shap.summary_plot(sv_fin, X_te15, plot_type="bar", show=False)
    plt.title(f"SHAP Global — {titulo_modelo}", pad=12)
    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_DIR, f"{nombre_archivo}_shap_global.png"), dpi=200)
    plt.show()

    idx = 0
    clase_pred = y_pred[idx]
    print(f"\nSHAP Local — Obs #{idx} → '{nombres_clases[clase_pred]}'")
    try:
        if es_binario:
            shap.plots.waterfall(expl_obj[idx], show=False)
        else:
            shap.plots.waterfall(expl_obj[idx, :, clase_pred], show=False)
        plt.tight_layout()
        plt.savefig(os.path.join(RESULTS_DIR, f"{nombre_archivo}_shap_waterfall_obs0.png"), dpi=200)
        plt.show()
    except Exception as e:
        print(f"No se pudo generar waterfall: {e}")

    return {
        'modelo': modelo_fin, 'top15': top15, 'metricas': metricas,
        'importancia_shap': df_imp, 'X_test_final': X_te15,
        'y_test': y_test, 'y_pred': y_pred, 'y_proba': y_proba
    }


# ══════════════════════════════════════════════════════════════
# BLOQUE 6 — MODELO A: 3 CLASES
# smoteenn_soft: ENN suave para evitar eliminación excesiva
# ══════════════════════════════════════════════════════════════

res_A = pipeline_completo_v5(
    X_raw=X_full,
    y=y_3_full,
    nombres_clases=nombres_3,
    titulo_modelo="MODELO A — Clasificación por Nivel de Riesgo (3 clases)",
    nombre_archivo="modelo_A_3clases",
    estrategia_balanceo='smoteenn_soft'   # ← suave para Modelo A
)

# ══════════════════════════════════════════════════════════════
# BLOQUE 7 — MODELO B: 2 CLASES
# smoteenn_strict
# ══════════════════════════════════════════════════════════════

res_B = pipeline_completo_v5(
    X_raw=X_full,
    y=y_2_full,
    nombres_clases=nombres_2,
    titulo_modelo="MODELO B — Identificación de Riesgo (2 clases)",
    nombre_archivo="modelo_B_binario",
    estrategia_balanceo='smoteenn_strict'  # ← estricto para Modelo B
)

# ══════════════════════════════════════════════════════════════
# BLOQUE 8 — RESUMEN COMPARATIVO
# ══════════════════════════════════════════════════════════════

metricas_A = res_A['metricas']
metricas_B = res_B['metricas']
resumen = pd.concat([metricas_A, metricas_B], ignore_index=True)

print("\n" + "="*80)
print("RESUMEN COMPARATIVO — MODELOS PROPIOS")
print("="*80)

display(resumen[[
    'modelo', 'accuracy', 'balanced_accuracy', 'f1_macro', 'f1_weighted',
    'auc', 'dummy_accuracy', 'dummy_balanced_accuracy',
    'n_train_raw', 'n_test', 'n_train_after_smoteenn'
]])

resumen.to_csv(os.path.join(RESULTS_DIR, "resumen_comparativo_modelos.csv"), index=False)

print("\nComparación orientativa con Zhang & Zhao (2026):")
print("- Rank classification: accuracy 91.0%, AUC 0.977")
print("- Risk identification: accuracy 97.2%, AUC 0.997")
print("\nNota: la comparación no es directa, porque este proyecto usa datos públicos, menor muestra,")
print("variables obtenidas vía yfinance y no dispone del panel empresa-año ni de todas las variables ESG cualitativas del estudio original.")
```
